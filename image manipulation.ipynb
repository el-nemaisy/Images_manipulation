### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Loading an image

```python
# We will load the image using matplotlib library

img = plt.imread("images\img1.jpeg")
```

### Get image information

```python
print(f"Image type: {type(img)}") # Print image type

print("============================")

print(f"Image shape: {img.shape}")  #print image shape

print("============================")

print(f"Image dim: {img.ndim}") # print image dimensions

print("============================")

print(f"Image size: {img.size}") # print image size (row * col * channel)

print("============================")

print(img) # print image as a martix
```

### Display the image

```python
# using matplotlib
plt.imshow(img )
plt.title("Sea View")
plt.axis("off")
plt.show()
```

## Color image

```python
# Red channel image
red_channel_image = np.zeros(img.shape , dtype=("int"))
red_channel_image[:,:,0]= img[:,:,0]
```

```python
plt.imshow(red_channel_image)
plt.title("Red channel")
plt.axis("off")
plt.show()
```

```python
# Blue channel image
blue_channel_img = np.zeros( img.shape , dtype = "int")
blue_channel_img[:,:,2] = img[:,:,2]
```

```python
plt.imshow(blue_channel_img)
plt.title("Blue Channel")
plt.axis("off")
plt.show()
```

```python
# Green channel image
green_channel_img = np.zeros(img.shape , dtype= "int")
green_channel_img[:,:,1]= img[:,:,1]
```

```python
plt.imshow(green_channel_img)
plt.axis("off")
plt.title("Green Channel")
plt.show()
```
